```python
import requests
result = requests.get("http://www.imdb.com/search/title?sort=num_votes,desc&start=1&title_type=feature&year=1950,2012")
```

```python
result
```

```python
import bs4
soup = bs4.BeautifulSoup(result.text,'lxml')
```

```python
cases= soup.find_all("div",class_="lister-item mode-advanced")
print(len(cases))
```

```python
year = []
Run =[]
name = []
director=[]
actors=[]
votes=[]
gross=[]
rating=[]
genre=[]

for i in cases:
    year_mov= i.find("span",class_="lister-item-year")
    year.append(year_mov.string)
    runtime_mov= i.find("span",class_="runtime")
    Run.append(runtime_mov.string)
    name_mov = i.find('img')['alt']
    name.append(name_mov)
    
    #print(name_mov)
    movies = i.find_all('p',class_="")
    #print(movies)
    direct_titles = [a.text for a in movies]
    direct=direct_titles[0].split('\n')[2:4]
    if direct[1]=='| ':
        direct.remove('| ')

    director.append(direct[:])
    

    
    actor_titles = [a.text for a in movies]
    actor=actor_titles[0].split('\n')[5:9]
    actors.append(actor)
    
    vote=i.find_all('p',class_='sort-num_votes-visible')
    
    n_votes=[name.text for name in vote]
    votescount=n_votes[0].split()[1]
    votes.append(votescount)
    
    n_gross=[name.text for name in vote]
    gross.append(n_gross[0].split()[-1])
    
    rat=i.find_all('div',class_='inline-block ratings-imdb-rating')
    n_rat=[strong.text for strong in rat]
    rating.append(n_rat[0].strip())
    
    gen=i.find('span',class_='genre')
    genre.append(gen.string.strip())
```

```python
import pandas as pd
df =pd.DataFrame({'Name':name,'Year':year,'Runtime':Run,'Director':director,'Actors':actors,'Votes':votes,'Gross':gross,'Rating':rating,'Genre':genre})
df
```
